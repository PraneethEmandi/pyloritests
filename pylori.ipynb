```python
import cv2
import numpy as np
import os

# Define a function to apply filters
def apply_filters(image_path, output_dir="filtered_images"):
    # Load the image
    image = cv2.imread(image_path)
    if image is None:
        print("Error: Image not found!")
        return
    
    # Ensure output directory exists
    os.makedirs(output_dir, exist_ok=True)
    
    # Apply different types of filters
    filters = {
        "original": image,
        "grayscale": cv2.cvtColor(image, cv2.COLOR_BGR2GRAY),
        "gaussian_blur": cv2.GaussianBlur(image, (15, 15), 0),
        "median_blur": cv2.medianBlur(image, 15),
        "bilateral_filter": cv2.bilateralFilter(image, 9, 75, 75),
        "edge_detection": cv2.Canny(image, 100, 200),
        "sharpening": cv2.filter2D(image, -1, np.array([[0, -1, 0], [-1, 5,-1], [0, -1, 0]])),
    }

    # Save all filtered images
    for filter_name, filtered_image in filters.items():
        # Convert grayscale images to BGR for saving
        if len(filtered_image.shape) == 2:
            filtered_image = cv2.cvtColor(filtered_image, cv2.COLOR_GRAY2BGR)
        
        # Save the filtered image
        output_path = os.path.join(output_dir, f"{filter_name}.jpg")
        cv2.imwrite(output_path, filtered_image)
        print(f"Saved {filter_name} filter to {output_path}")
    
    print(f"All filtered images saved in '{output_dir}' directory.")

# Example usage
image_path = "0004.jpg"  # Replace with your image path
apply_filters(image_path)
```

```python
import cv2
import numpy as np
import os

def apply_extended_filters(image_path, output_dir="filtered_images3"):
    # Load the image
    image = cv2.imread(image_path)
    if image is None:
        print("Error: Image not found!")
        return

    # Ensure output directory exists
    os.makedirs(output_dir, exist_ok=True)

    # Grayscale conversion
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply different filters
    filters = {
        "original": image,
        "grayscale": gray_image,
        "box_blur": cv2.boxFilter(image, -1, (5, 5)),
        "gaussian_blur": cv2.GaussianBlur(image, (15, 15), 0),
        "median_blur": cv2.medianBlur(image, 15),
        "bilateral_filter": cv2.bilateralFilter(image, 9, 75, 75),
        "sobel_x": cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=5),
        "sobel_y": cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=5),
        "laplacian": cv2.Laplacian(gray_image, cv2.CV_64F),
        "edge_detection": cv2.Canny(image, 100, 200),
        "sharpening": cv2.filter2D(image, -1, np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])),
    }

    # Process and save each filtered image
    for filter_name, filtered_image in filters.items():
        # Normalize Sobel and Laplacian results to 8-bit if needed
        if isinstance(filtered_image, np.ndarray) and filtered_image.dtype == np.float64:
            filtered_image = cv2.convertScaleAbs(filtered_image)
        
        # Convert grayscale images to BGR for saving
        if len(filtered_image.shape) == 2:
            filtered_image = cv2.cvtColor(filtered_image, cv2.COLOR_GRAY2BGR)
        
        # Save the filtered image
        output_path = os.path.join(output_dir, f"{filter_name}.jpg")
        cv2.imwrite(output_path, filtered_image)
        print(f"Saved {filter_name} filter to {output_path}")
    
    print(f"All filtered images saved in '{output_dir}' directory.")

# Example usage
image_path = "0004.jpg"  # Replace with your image path
apply_extended_filters(image_path)
```

```python
import cv2
import numpy as np
import os
import math

def apply_all_filters(image_path, output_dir="all_filters"):
    # Load the image
    image = cv2.imread(image_path)
    if image is None:
        print("Error: Image not found!")
        return
    
    # Ensure output directory exists
    os.makedirs(output_dir, exist_ok=True)

    # Grayscale version
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Filters and effects dictionary
    filters = {}

    # Gabor Filter
    kernel = cv2.getGaborKernel((21, 21), 5, np.pi / 4, 10, 0.5, 0, ktype=cv2.CV_32F)
    filters["gabor"] = cv2.filter2D(image, cv2.CV_8UC3, kernel)

    # Histogram Equalization
    ycrcb = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
    ycrcb[:, :, 0] = cv2.equalizeHist(ycrcb[:, :, 0])
    filters["histogram_equalization"] = cv2.cvtColor(ycrcb, cv2.COLOR_YCrCb2BGR)

    # Cartoon Effect
    gray = cv2.medianBlur(gray_image, 7)
    edges = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                  cv2.THRESH_BINARY, 9, 10)
    color = cv2.bilateralFilter(image, 9, 300, 300)
    filters["cartoon"] = cv2.bitwise_and(color, color, mask=edges)

    # CLAHE (Contrast Limited Adaptive Histogram Equalization)
    lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
    lab[:, :, 0] = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8)).apply(lab[:, :, 0])
    filters["clahe"] = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)

    # Sepia Effect
    sepia_kernel = np.array([[0.272, 0.534, 0.131],
                             [0.349, 0.686, 0.168],
                             [0.393, 0.769, 0.189]])
    sepia = cv2.transform(image, sepia_kernel)
    filters["sepia"] = np.clip(sepia, 0, 255).astype(np.uint8)

    # Invert Filter
    filters["invert"] = cv2.bitwise_not(image)

    # Warm Filter
    filters["warm"] = cv2.addWeighted(image, 0.8, np.full_like(image, (0, 50, 100)), 0.2, 0)

    # Cool Filter
    filters["cool"] = cv2.addWeighted(image, 0.8, np.full_like(image, (100, 50, 0)), 0.2, 0)

    # Emboss Effect
    emboss_kernel = np.array([[-2, -1, 0],
                               [-1, 1, 1],
                               [0, 1, 2]])
    filters["emboss"] = cv2.filter2D(image, -1, emboss_kernel) + 128

    # Motion Blur
    kernel = np.zeros((15, 15))
    kernel[int((15 - 1) / 2), :] = np.ones(15) / 15
    filters["motion_blur"] = cv2.filter2D(image, -1, kernel)

    # Frosted Glass Effect
    frosted = np.zeros_like(image)
    h, w, _ = image.shape
    for i in range(h):
        for j in range(w):
            offset_x = np.random.randint(-2, 3)
            offset_y = np.random.randint(-2, 3)
            x = min(w - 1, max(0, j + offset_x))
            y = min(h - 1, max(0, i + offset_y))
            frosted[i, j] = image[y, x]
    filters["frosted_glass"] = frosted

    # Vignette Effect
    # rows, cols = image.shape[:2]
    # X_kernel = cv2.getGaussianKernel(cols, 200)
    # Y_kernel = cv2.getGaussianKernel(rows, 200)
    # mask = 255 * (X_kernel * Y_kernel.T) / np.linalg.norm(X_kernel * Y_kernel.T)
    # vignette = np.copy(image)
    # for i in range(3):
    #     vignette[:, :, i] = vignette[:, :, i] * mask
    # filters["vignette"] = vignette

    # Vignette Effect
    # rows, cols = image.shape[:2]
    # X_kernel = cv2.getGaussianKernel(cols, 200)
    # Y_kernel = cv2.getGaussianKernel(rows, 200)
    # mask = 255 * (X_kernel * Y_kernel.T) / np.linalg.norm(X_kernel * Y_kernel.T)
    # mask = mask[:, :, np.newaxis]  # Add a third dimension to match the image shape
    # vignette = np.copy(image)
    # vignette = vignette * mask.astype(image.dtype)  # Apply mask to all channels
    # filters["vignette"] = vignette


    # HDR Effect
    filters["hdr"] = cv2.detailEnhance(image, sigma_s=12, sigma_r=0.15)

    # Sketch Effect
    inv_gray = cv2.bitwise_not(gray_image)
    blurred = cv2.GaussianBlur(inv_gray, (21, 21), 0)
    filters["sketch"] = cv2.divide(gray_image, 255 - blurred, scale=256)

    # Watercolor Effect
    filters["watercolor"] = cv2.stylization(image, sigma_s=60, sigma_r=0.6)

    # Box Filter
    filters["box_filter"] = cv2.boxFilter(image, -1, (5, 5))

    # Sobel Filters
    filters["sobel_x"] = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=5)
    filters["sobel_y"] = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=5)

    # Laplacian Filter
    filters["laplacian"] = cv2.Laplacian(gray_image, cv2.CV_64F)

    # Scharr Filter
    filters["scharr_x"] = cv2.Scharr(gray_image, cv2.CV_64F, 1, 0)
    filters["scharr_y"] = cv2.Scharr(gray_image, cv2.CV_64F, 0, 1)

    # Unsharp Masking
    gaussian_blur = cv2.GaussianBlur(image, (9, 9), 10.0)
    filters["unsharp_mask"] = cv2.addWeighted(image, 1.5, gaussian_blur, -0.5, 0)

    # Morphological Filters
    kernel = np.ones((5, 5), np.uint8)
    filters["erosion"] = cv2.erode(image, kernel, iterations=1)
    filters["dilation"] = cv2.dilate(image, kernel, iterations=1)
    filters["gradient"] = cv2.morphologyEx(image, cv2.MORPH_GRADIENT, kernel)

    # Thresholding
    _, binary = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
    filters["threshold_binary"] = binary

    _, otsu = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    filters["threshold_otsu"] = otsu

    # Adaptive Thresholding
    adaptive = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                     cv2.THRESH_BINARY, 11, 2)
    filters["adaptive_threshold"] = adaptive

    # Save all filters
    for name, img in filters.items():
        if isinstance(img, np.ndarray) and img.dtype != np.uint8:
            img = cv2.convertScaleAbs(img)
        if len(img.shape) == 2:
            img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
        cv2.imwrite(os.path.join(output_dir, f"{name}.jpg"), img)
        print(f"Saved {name}")

# Example usage
apply_all_filters("0004.jpg")
```
